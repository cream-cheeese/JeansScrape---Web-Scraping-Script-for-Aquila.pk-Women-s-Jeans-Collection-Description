```python
import requests
from bs4 import BeautifulSoup
import csv

# Send a GET request to the website and create a BeautifulSoup object
url = 'https://aquila.pk/collections/women-jeans?filter.v.availability=1&filter.v.option.size=30&filter.v.price.gte=0&filter.v.price.lte=2140'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

# Find all the items within the container
items = soup.select('.grid-item.col-6.col-md-4')

# Create a data structure to store the extracted information
data = []

for item in items:
    # Retrieve item details
    item_details = item.select_one('.product-details')
    name = item_details.a.span.text.replace('\n', '')

    # Retrieve price details
    item_pricebox = item.select_one('.price-sale')
    old_price_element = item_pricebox.select_one('.old-price') if item_pricebox else item.select_one('.price-regular')
    sale_price_element = item_pricebox.select_one('.special-price') if item_pricebox else None

    # Extract price values
    old_price = old_price_element.text.replace('\n', '') if old_price_element else 'N/A'
    sale_price = sale_price_element.text if sale_price_element else 'N/A'

    href = item_details.a['href']

    # Store the extracted data in a dictionary
    item_data = {
        'name': name,
        'old_price': old_price,
        'sale_price': sale_price,
        'link': 'https://aquila.pk/' + href
    }

    # Add the item data to the list
    data.append(item_data)

# Store the extracted data in a file (CSV)
filename = 'aquila_products.csv'
fields = ['name', 'old_price', 'sale_price', 'link']

with open(filename, 'w', newline='') as csvfile:
    # Create a CSV writer and write the header
    writer = csv.DictWriter(csvfile, fieldnames=fields)
    writer.writeheader()
    
    # Write the data rows
    writer.writerows(data)

# Print the extracted data
for item in data:
    print('Name:', item['name'])
    print('Price:', item['old_price'])
    print('Sale Price:', item['sale_price'])
    print('Link:', item['link'])
    print()
```
